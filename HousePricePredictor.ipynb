```python
## importing the necessary libs
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing the dataset (https://www.kaggle.com/datasets/anmolkumar/house-price-prediction-challenge)
df = pd.read_csv("train.csv") #reading the dataset
df.sample(10)
```

```python
#cleaning and refining data
df['BHK_OR_RK'] = df['BHK_OR_RK'].replace('BHK', 1)
df['BHK_OR_RK'] = df['BHK_OR_RK'].replace('RK', 2)
df.sample()
```

```python
#importing decision tree

model = DecisionTreeRegressor()
X = df.drop(['PRICE_IN_LACS'], axis = 1) #features
y = df['PRICE_IN_LACS'] #target or dependant variable
```

```python
#testing accuracy for the Machine Learning Model using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

model.fit(X_train, y_train)
pre = model.predict(X_test)
accuracy = r2_score(y_test, pre)
accuracy
```
